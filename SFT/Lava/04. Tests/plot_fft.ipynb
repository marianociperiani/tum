```python
import numpy as np
import matplotlib.pyplot as plt

# data read
with open("256-samples.txt", "r") as file:
    data = [float(value) for value in file.read().split()]


# Convert the data to a NumPy array
in_data = np.array(data)

# Perform the FFT on in_data
fft_result = np.fft.fft(in_data)
frequencies = np.fft.fftfreq(len(in_data))

# Calculate the magnitude of the FFT result
fft_magnitude = np.abs(fft_result)
print(fft_magnitude)
# Create a frequency axis for plotting
sampling_rate = 1  # Adjust this if you have a different sampling rate
frequency_axis = np.fft.fftfreq(len(in_data), 1 / sampling_rate)

# Plot the magnitude of the FFT result
plt.figure(figsize=(10, 6))
plt.plot(fft_magnitude)
plt.title('FFT Magnitude')
plt.grid(True)

# Show the plot
plt.show()
```
